```python
"""
This perfomrs BFS starting with words provided by get_related_words method
goes level by level searching on wikipedia related nouns to the parent word


"""

from queue import Queue


## max depth for BFS
MAX_LEVELS = 1

q = Queue()

pedestrian_words = get_related_words("pedestrian");


# initialize queue with starting words
for word in pedestrian_words:
    q.put(word)

# visited words set so as to not form cycles
visited_words = pedestrian_words;

print("STARTING WITH = ", pedestrian_words)

level = 0


while level <= MAX_LEVELS and not q.empty():

    print("LEVEL = " + str(level))
    for i in range(q.qsize()): 

        word = q.get()
        print("ROOT = " + word)

        phrase_list = get_phrases_from_wiki(word)
        ending_noun_list = get_ending_noun_list(phrase_list)
        for ending_noun in ending_noun_list:
            if len(ending_noun) != 0 and ending_noun not in visited_words:
                print(ending_noun + " :: ", end='')
                q.put(ending_noun.lower())
                visited_words.add(ending_noun)

        print('\n')

    print("\n\n\nvisited\n", visited_words)
    print('=========================LEVEL END =========================\n\n\n')
    level += 1
    
```
